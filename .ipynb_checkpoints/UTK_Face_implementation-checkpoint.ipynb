#### Stage A – Dataset & Client Simulation

* Load UTKFace

* Preprocess images + gender labels

* Split dataset across N clients (IID first)

* Each client has its own local dataset

#### Stage B – Clean Federated Learning (No Poisoning)

* CNN model for gender classification

* Federated training using FedAvg (baseline)

#### Track:

* Global accuracy

* Per-client accuracy

* Fairness metrics

#### Stage C – Malicious Clients (Task 2)

* Label-flipping or backdoor attacks

* Compare accuracy + fairness degradation

```python
import os
import random
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

from PIL import Image
import flwr as fl
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(DEVICE)
```

```python
# -------------------------------
# Dataset
# -------------------------------
class UTKFaceDataset(Dataset):
    def __init__(self, root_dir, image_paths, transform=None):
        self.root_dir = root_dir
        self.image_paths = image_paths
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_name = self.image_paths[idx]
        path = os.path.join(self.root_dir, img_name)

        try:
            img = Image.open(path).convert("RGB")
        except:
            return self.__getitem__((idx + 1) % len(self.image_paths))

        gender = int(img_name.split("_")[1])  # 0=male, 1=female

        if self.transform:
            img = self.transform(img)

        return img, gender


# -------------------------------
# CNN Model
# -------------------------------
class GenderCNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 32, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(32, 64, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
        )

        self.classifier = nn.Sequential(
            nn.Linear(64 * 16 * 16, 128),
            nn.ReLU(),
            nn.Linear(128, 2)
        )

    def forward(self, x):
        x = self.features(x)
        x = x.view(x.size(0), -1)
        return self.classifier(x)

DATA_DIR = "UTKFace"
NUM_CLIENTS = 10
# -------------------------------
# Load Data
# -------------------------------
all_images = [
    f for f in os.listdir(DATA_DIR)
    if f.endswith(".jpg") or f.endswith(".jpg.chip.jpg")
]

random.shuffle(all_images)

train_imgs, test_imgs = train_test_split(
    all_images, test_size=0.2, random_state=42
)

def split_clients(data, n):
    np.random.shuffle(data)
    return np.array_split(data, n)

client_splits = split_clients(train_imgs, NUM_CLIENTS)

transform = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize([0.5]*3, [0.5]*3)
])

def make_loader(imgs, shuffle=True):
    ds = UTKFaceDataset(DATA_DIR, imgs, transform)
    return DataLoader(ds, batch_size=BATCH_SIZE, shuffle=shuffle)

test_loader = make_loader(test_imgs, shuffle=False)

# -------------------------------
# Train / Eval
# -------------------------------
def train(model, loader):
    model.train()
    opt = torch.optim.Adam(model.parameters(), lr=1e-3)
    loss_fn = nn.CrossEntropyLoss()

    for x, y in loader:
        x, y = x.to(DEVICE), y.to(DEVICE)
        opt.zero_grad()
        loss = loss_fn(model(x), y)
        loss.backward()
        opt.step()

def evaluate(model, loader):
    model.eval()
    correct, total = 0, 0
    preds, labels = [], []

    with torch.no_grad():
        for x, y in loader:
            x, y = x.to(DEVICE), y.to(DEVICE)
            out = model(x)
            p = out.argmax(dim=1)
            correct += (p == y).sum().item()
            total += y.size(0)
            preds.extend(p.cpu().numpy())
            labels.extend(y.cpu().numpy())

    return correct / total, np.array(preds), np.array(labels)

# -------------------------------
# Fairness Metrics
# -------------------------------
def fairness_metrics(preds, labels):
    # SPD
    male_rate = preds[labels == 0].mean()
    female_rate = preds[labels == 1].mean()
    spd = male_rate - female_rate

    # EOD
    cm = confusion_matrix(labels, preds)
    tpr_male = cm[0, 0] / (cm[0, 0] + cm[0, 1])
    tpr_female = cm[1, 1] / (cm[1, 0] + cm[1, 1])
    eod = tpr_male - tpr_female

    return spd, eod

# -------------------------------
# Flower Client
# -------------------------------
class FLClient(fl.client.NumPyClient):
    def __init__(self, loader):
        self.model = GenderCNN().to(DEVICE)
        self.loader = loader

    def get_parameters(self, config):
        return [v.cpu().numpy() for v in self.model.state_dict().values()]

    def set_parameters(self, parameters):
        keys = self.model.state_dict().keys()
        state = {k: torch.tensor(v) for k, v in zip(keys, parameters)}
        self.model.load_state_dict(state)

    def fit(self, parameters, config):
        self.set_parameters(parameters)
        train(self.model, self.loader)
        return self.get_parameters(config), len(self.loader.dataset), {}

    def evaluate(self, parameters, config):
        self.set_parameters(parameters)
        acc, _, _ = evaluate(self.model, test_loader)
        return float(acc), len(test_loader.dataset), {"accuracy": acc}

def client_fn(cid):
    return FLClient(make_loader(client_splits[int(cid)]))

# -------------------------------
# Run FL
# -------------------------------
acc_history = []

class Strategy(fl.server.strategy.FedAvg):
    def aggregate_evaluate(self, rnd, results, failures):
        acc = np.mean([r.metrics["accuracy"] for _, r in results])
        acc_history.append(acc)
        return super().aggregate_evaluate(rnd, results, failures)

strategy = Strategy(
    fraction_fit=1.0,
    min_fit_clients=NUM_CLIENTS,
    min_available_clients=NUM_CLIENTS,
)

fl.simulation.start_simulation(
    client_fn=client_fn,
    num_clients=NUM_CLIENTS,
    config=fl.server.ServerConfig(num_rounds=ROUNDS),
    strategy=strategy,
)

# -------------------------------
# Final Evaluation
# -------------------------------
global_model = GenderCNN().to(DEVICE)
params = strategy.initial_parameters
keys = global_model.state_dict().keys()
state = {k: torch.tensor(v) for k, v in zip(keys, params)}
global_model.load_state_dict(state)

acc, preds, labels = evaluate(global_model, test_loader)
spd, eod = fairness_metrics(preds, labels)

print(f"\nFinal Accuracy: {acc:.4f}")
print(f"Statistical Parity Difference: {spd:.4f}")
print(f"Equal Opportunity Difference: {eod:.4f}")

# -------------------------------
# Plots
# -------------------------------
plt.plot(acc_history)
plt.xlabel("Rounds")
plt.ylabel("Accuracy")
plt.title("Federated Learning Accuracy (Clean)")
plt.show()
```
